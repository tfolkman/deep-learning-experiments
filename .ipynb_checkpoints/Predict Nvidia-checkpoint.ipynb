```python
import pandas_datareader as pdr
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, LSTM, Bidirectional, Dropout
import pandas as pd
from numpy import log
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nvidia_data = pdr.get_data_google("nvda")
nvidia_data.head(6)
```

```python
ret = lambda x,y: log(y/x) #Log return 
zscore = lambda x:(x -x.mean())/x.std() # zscore
```

```python
Res = pd.DataFrame()
Res['o_2_c'] = zscore(ret(nvidia_data.Open,nvidia_data.Close))
Res['o_2_h'] = zscore(ret(nvidia_data.Open,nvidia_data.High))
Res['o_2_l'] = zscore(ret(nvidia_data.Open,nvidia_data.Low))
Res['h_2_c'] = zscore(ret(nvidia_data.High,nvidia_data.Close))
Res['h_2_l'] = zscore(ret(nvidia_data.High,nvidia_data.Low))
Res['vol'] = zscore(nvidia_data.Volume)
Res['c1_c0'] = ret(nvidia_data.Close,nvidia_data.Close.shift(-1)) #Tommorows return 
```

```python
Res.head(9)
```

```python
Res.c1_c0.plot()
```

```python
data_dim = 6
timesteps = 7
batch_size = 32
```

```python
def format_data(data, timesteps):
    shape = data.shape
    X = []
    y = []
    for i in range(shape[0]-timesteps+1):
        X.append(data[i:i+timesteps, :-1])
        y.append(data[i+timesteps-1, -1])
    return np.array(X), np.array(y)
```

```python
res_values = Res.values
X, y = format_data(res_values, timesteps)
```

```python
n_points = X.shape[0]
train_end = int(int(n_points * 0.8) / batch_size) * batch_size
val_end = int(int(n_points * 0.9) / batch_size) * batch_size
test_end = int(n_points / batch_size) * batch_size

X_train = X[:train_end]
y_train = y[:train_end]
X_val = X[train_end:val_end]
y_val = y[train_end:val_end]
X_test = X[val_end:test_end]
y_test = y[val_end:test_end]
```

```python
pd.Series(y_test).plot()
```

```python
model = Sequential()
model.add(LSTM(128, return_sequences=True, stateful=False,batch_input_shape=(batch_size, timesteps, data_dim)))
model.add(LSTM(64, return_sequences=True, stateful=False))
model.add(LSTM(32, stateful=False))
model.add(Dense(1, activation='linear'))

model.compile(loss='mse',
              optimizer='adam',
              metrics=['mse'])
```

```python
model.fit(X_train, y_train,
          batch_size=batch_size, epochs=3, shuffle=False,
          validation_data=(X_val, y_val))
```

```python
predicted_val = model.predict(X_val)
x_val = list(range(len(predicted_val)))
```

```python
plt.plot(x_test, y_test, label='truth')
plt.plot(x_val, predicted_val, label='predicted')
plt.title("val set")
plt.legend()
```

```python
predicted_test = model.predict(X_test)
x_test = list(range(len(predicted_test)))
```

```python
plt.plot(x_test, y_test, label='truth')
plt.plot(x_test, predicted_test, label='predicted')
plt.title("Test Set")
plt.legend()
```

```python
predicted_train = model.predict(X_train)
x_train = list(range(len(predicted_train)))
```

```python
plt.plot(x_train, y_train, label='truth')
plt.plot(x_train, predicted_train, label='predicted')
plt.title("train set")
plt.legend()
```
